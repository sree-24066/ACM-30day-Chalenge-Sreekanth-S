```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")
sns.set(style="whitegrid")
pd.set_option('display.max_columns', None)

df = pd.read_csv("mental_health_workplace_survey.csv")  # Use your uploaded file
df.head()
```

```python
print("Total Rows:", df.shape[0])
print("Total Columns:", df.shape[1])

print("\n Dataset Info:")
df.info()

print("\n Summary Statistics:")
df.describe()
```

```python
# Boxplots to visualize outliers
features_to_check = ['WorkHoursPerWeek', 'SleepHours', 'StressLevel']

for col in features_to_check:
    plt.figure(figsize=(6, 1.5))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot for {col}")
    plt.show()
```

```python
# Function to remove outliers using IQR
def remove_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    original_shape = df.shape[0]
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    removed = original_shape - df.shape[0]
    print(f"{removed} outliers removed from '{col}'")
    return df

# Apply to selected features
for feature in ['WorkHoursPerWeek', 'SleepHours', 'StressLevel']:
    df = remove_outliers_iqr(df, feature)
```

```python
# 1. Burnout Distribution
sns.countplot(x='BurnoutRisk', data=df)
plt.title("Burnout Distribution")
plt.xlabel("Burnout Risk (1 = Yes, 0 = No)")
plt.ylabel("Count")
plt.show()

# 2. Sleep Hours vs Burnout
sns.boxplot(x='BurnoutRisk', y='SleepHours', data=df)
plt.title("Sleep Hours by Burnout Risk")
plt.xlabel("Burnout Risk (1 = Yes, 0 = No)")
plt.ylabel("Sleep Hours")
plt.show()

# 3. Scatterplot: Stress vs Sleep (colored by Burnout)
sns.scatterplot(x='SleepHours', y='StressLevel', hue='BurnoutRisk', data=df, palette='coolwarm')
plt.title("Stress vs Sleep (Colored by Burnout Risk)")
plt.xlabel("Sleep Hours")
plt.ylabel("Stress Level")
plt.legend(title="Burnout Risk")
plt.show()

# 4. Correlation Heatmap (numeric only)
plt.figure(figsize=(12, 6))
numeric_df = df.select_dtypes(include=['int64', 'float64'])  # Only keep numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```
